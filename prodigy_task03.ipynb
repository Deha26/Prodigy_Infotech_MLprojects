IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import warnings
warnings.filterwarnings("ignore")
import os
from skimage.transform import resize
from skimage.io import imread
from tqdm import tqdm
```

```python
train_p="C:\\Users\\Dell\\Downloads\\dog-cat\\train"
test_p="C:\\Users\\Dell\\Downloads\\dog-cat\\test"
```

LOAD IMAGE DATASET

```python
train_data,train_tar=[],[]
for dir in tqdm(os.listdir(train_p)):
    path=os.path.join(train_p+"\\"+dir)
    for img in tqdm(os.listdir(path)):
        img_r=imread(os.path.join(path+"\\"+img))
        img_res=resize(img_r,(50,50,3))
        img_nor=img_res/255.0
        train_data.append(img_nor.flatten())
        train_tar.append(dir)
train_data=np.array(train_data)
train_tar=np.array(train_tar)
```

```python
train_data.shape
```

```python
train_tar
```

LABEL ENCODING 

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
train_le=le.fit_transform(train_tar)
train_le
```

```python
test_data,test_tar=[],[]
for dir in tqdm(os.listdir(test_p)):
    path=os.path.join(test_p+"\\"+dir)
    for img in tqdm(os.listdir(path)):
        img_r=imread(os.path.join(path+"\\"+img))
        img_res=resize(img_r,(50,50,3))
        img_nor=img_res/255.0
        test_data.append(img_nor.flatten())
        test_tar.append(dir)
test_data=np.array(test_data)
test_tar=np.array(test_tar)
```

```python
test_data.shape
```

```python
test_tar
```

```python
test_le=le.transform(test_tar)
test_le
```

CREATING SVM MODEL

```python
from sklearn.svm import SVC
model=SVC()   # rbf kernel by default
model.fit(train_data,train_le)
```

```python
#train accuracy
model.score(train_data,train_le)
```

```python
#test accuracy
model.score(test_data,test_le)
```

STANDARDISING THE DATA

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
train_scaled=sc.fit_transform(train_data)
test_scaled=sc.fit_transform(test_data)
```

```python
model.fit(train_scaled,train_le)
```

```python
#train accuracy
model.score(train_scaled,train_le)
```

```python
#test sccuracy
model.score(test_scaled,test_le)
```
